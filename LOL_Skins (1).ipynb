```python
hero_id = []
hero_name_title = []
hero_dict={}
hero_id = hero_dict.keys()
hero_skin = {}
```

```python
url = "https://game.gtimg.cn/images/lol/act/img/js/heroList/hero_list.js?v=23"
response = requests.get(url)
text = response.text.encode('utf-8')
herolist = json.loads(text)['hero']
```

```python
for hero in herolist:
    #print(hero)
    #hero_id.append(hero['heroId'])
    #hero_name_title.append(hero['name']+"-"+hero['title'])
    hero_dict[hero['heroId']]=hero['name']+"-"+hero['title']
```

```python
# create another dictionary where hero's title and name are keys while the context of the map are the list of urls of the heros' skins
for id in hero_id:
    url = "https://game.gtimg.cn/images/lol/act/img/js/hero/"+id+".js"
    response = requests.get(url)
    temp = response.text.encode()
    data = json.loads(temp)
    
    skinlist = []
    print("正在爬取"+hero_dict[id]+"的皮肤")
    for mainskin in data["skins"]:
        if (mainskin["mainImg"] != ''):
            skinlist.append(mainskin["mainImg"])
    print("complete")
    hero_skin[hero_dict[id]] = skinlist
```

```python
#make directiory 
def mkdir(name):
    #os.chdir('/Users/apple/Desktop/Skins')
    
    '''
    I created a folder named Skins on my desktop in advance,
    you need to change the following line to your own path
    in order to let the program run properly 
    '''
    
    ########change your path here!!!!##############
    
    path = '/Users/apple/Desktop/Skins/'+name    
    
    ###############################################
    folder = os.path.exists(path)
 
    if not folder:                   #check if the folder exists, if it doesn't exist
        os.makedirs(path)            #makedirs on the given path
        print ("---  new folder" + name + "creating" + "...")
        print ("---  OK  ---")
    
    return path
```

```python
# Store the corresponding skins in the dictionary built into the new repositary created
def store_data_under_dir(id,path):
    os.chdir(path)
    count = 1
    for temp_url in hero_skin[hero_dict[id]]:
        print("正在爬取"+hero_dict[id]+"的皮肤")
        r = requests.get(temp_url)
        file = hero_dict[str(id)] + str(count) +'.jpeg'
        with open(file,'wb') as f:
            f.write(r.content)
        count+=1
        print("complete")
```

```python
def store_skins(data):      
    #os.chdir('/Users/apple/Desktop/Skins')
    for temp in data:
        hero_name = data[temp]
        #print(hero_name)
        store_data_under_dir(temp,mkdir(hero_name))
```

```python
store_skins(hero_dict)

'''
you can now check the folder on the desktop now,
if every goes well,
heros'names folders are created, with their all skins inside
'''
```
